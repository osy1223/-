```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = [ [2,81], [4, 93], [6, 91], [8,97] ]

x = [i[0] for i in data]
y = [i[1] for i in data]
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y, color='red')
plt.show()
```

```python
x_data = np.array(x)
y_data = np.array(y)

a = 0
b = 0

lr = 0.05

epochs = 2001


for i in range(epochs):
    y_pred = a * x_data + b
    error = y_data - y_pred 
    
    a_diff = - (1/len(x_data)) * sum(x_data * (error))
    b_diff = - (1/len(x_data)) * sum(y_data - y_pred)
    
    a = a - lr * a_diff
    b = b - lr * b_diff    
    
    if i % 100 == 0:        
        print("epoch=%.f, 기울기=%.04f, 절편=%.04f" % (i, a, b))
```

```python
y_pred = a * x_data + b
plt.scatter(x,y, color='red')
plt.plot( [min(x_data), max(x_data)] , [min(y_pred), max(y_pred)] , color='green')
plt.show()
```
