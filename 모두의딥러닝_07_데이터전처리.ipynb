```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), 
            linewidths = 0.1,
            vmax = 0.5,
            cmap=plt.cm.gist_heat,
            linecolor='white',
            annot=True)
```
